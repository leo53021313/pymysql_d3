```python
import pymysql
from configparser import ConfigParser

config = ConfigParser()
config.read('../Chapter1/config.ini') # 指定設定檔的檔案路徑

connection = pymysql.connect(
    host=config.get('DB', 'host'),
    user=config.get('DB', 'user'),
    password=config.get('DB', 'password'),
    port=config.getint('DB', 'port'),
    cursorclass=pymysql.cursors.DictCursor,
)

print(connection.open)
```

建立資料庫

```python
database = "chapter2"
with connection.cursor() as cursor:
    sql = f"""
        CREATE DATABASE IF NOT EXISTS {database}
    """
    # 執行建立的 SQL 語句
    cursor.execute(sql)

    # 執行查看資料庫
    cursor.execute('SHOW DATABASES;')
    dbs = cursor.fetchall()

print(dbs)

```

建立資料表

```python
connection = pymysql.connect(
    host=config.get('DB', 'host'),
    user=config.get('DB', 'user'),
    password=config.get('DB', 'password'),
    port=config.getint('DB', 'port'),
    cursorclass=pymysql.cursors.DictCursor,
    database=database,
)

""" 建立 user 資料表
    id 主鍵
    name 字串 不能為空
    age 整數 
    username 字串 不能為空 必須唯一
    password 字串 不能為空
"""
with connection.cursor() as cursor:
    sql = """
        CREATE TABLE IF NOT EXISTS user (
            id INT AUTO_INCREMENT PRIMARY KEY,
            name VARCHAR(255) NOT NULL,
            age INT,
            username VARCHAR(255) NOT NULL UNIQUE,
            password VARCHAR(255) NOT NULL
        )
    """
    cursor.execute(sql)

    cursor.execute("SHOW TABLES;")
    tables = cursor.fetchall()
    print(tables)
```

寫入資料

```python
from pprint import pprint
with connection.cursor() as cursor:
    sql = """
        INSERT INTO user (name, age, username, password)
        VALUES ('Pochita', 10, 'pochita_10', 123456)
    """
    # 執行寫入的 SQL 語句
    cursor.execute(sql)
    
    # 執行查詢資料表
    cursor.execute('SELECT * from user;')

    # 取得查詢的所有資料
    users = cursor.fetchall()
    print(users)
```

更新資料

```python
with connection.cursor() as cursor:
    sql = """
        UPDATE user SET age = 22 WHERE username = 'pochita_10'
    """
    cursor.execute(sql)
    
    # 可以搭配 rowcount 屬性來檢查受影響的行數
    # 提交資料庫的變更
    print(cursor.rowcount)
    if cursor.rowcount == 1:
        # 提交資料庫的變更
        connection.commit()
    else:
        # 如果不是一筆的變動則回朔
        connection.rollback()


    cursor.execute("SELECT * FROM user;")
    result = cursor.fetchall()

pprint(result)
```

```python
import pymysql
from configparser import ConfigParser

config = ConfigParser()
config.read('../Chapter1/config.ini') # 指定設定檔的檔案路徑

connection = pymysql.connect(
    host=config.get('DB', 'host'),
    user=config.get('DB', 'user'),
    password=config.get('DB', 'password'),
    port=config.getint('DB', 'port'),
    cursorclass=pymysql.cursors.DictCursor,
    database='chapter2'
)

with connection.cursor() as cursor:
    sql = """
        DROP TABLES IF EXISTS user;
    """
    cursor.execute(sql)
    cursor.execute("SHOW TABLES;")
    result = cursor.fetchall()

print(result)
```
