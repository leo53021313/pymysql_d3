使用設定檔建立資料庫連線

```python
import pymysql
from configparser import ConfigParser
# 讀取 config.ini 檔案取得資料庫連線資訊
config = ConfigParser()
config.read('config.ini')

# 建立資料庫連線
connection = pymysql.connect(
    host=config['DB']['host'],
    user=config['DB']['user'],
    password=config['DB']['password'],
    port=int(config['DB']['port']),
    cursorclass=pymysql.cursors.DictCursor
)

with connection.cursor() as cursor:
    query = """
        SHOW DATABASES;
    """
    cursor.execute(query)
    result = cursor.fetchall()
print(result)
```

```python
from sql_utils import SqlManager
from pprint import pprint

db = SqlManager()

result = db.sql_query(SqlManager.full_passengers)
#print(result)
print(result[0].keys())

leonard_male = []
for r in result:
    if "Leonard" in r['pname'] and r['sex'] == 'male' and r['pname'].startswith('Leonard') is False:
        leonard_male.append({
            'id': r['id'],
            'pclass': r['pclass'],
            'pname': r['pname']
        })

print(len(leonard_male))
pprint(leonard_male, sort_dicts=False)
```
