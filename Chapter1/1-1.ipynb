建立資料庫連線

```python
import pymysql
# 建立 MySQL 資料庫連線
# connect 函式回傳 connection 物件
connection = pymysql.connect(
    # MySQL 伺服器的 IP 位址
    host = 'localhost',
    # 登入 MySQL 的使用者名稱
    user = 'root',
    # 登入 MySQL 的密碼
    password = 'P@ssw0rd',
    # MySQL 伺服器的端口
    port = 3306
)

print(connection)
```

```python
from pprint import pprint

# 建立游標
cursor = connection.cursor()
print(cursor)

# 使用游標執行 SQL 語句
cursor.execute('SHOW DATABASES;')

# 取得所有資料
dbs = cursor.fetchall()

# 輸出查詢結果
print(dbs)

# 關閉游標
cursor.close()

# 關閉資料庫連線
connection.close()

print(connection.open)
```

建議的寫法

```python
connection = pymysql.connect(
    host='localhost',
    user='root', 
    password='P@ssw0rd',
    port=3306, 
)

# 建立游標
with connection.cursor() as Cursor:
    # 使用游標執行 SQL 語句
    sql = """
        SHOW DATABASES;
    """
    Cursor.execute(sql)

    # 取得所有資料
    databases = Cursor.fetchall()
# 使用 with 自動關閉 cursor + connection
# 因為有設定 __enter__ __exit___

```

```python
from pprint import pprint

# 使用 DB > TABLE 查詢
with connection.cursor() as Cursor:
    # 使用游標執行 SQL 語句
    sql = """
        SHOW TABLES FROM my_titanic;
    """
    Cursor.execute(sql)

    # 取得所有資料
    result = Cursor.fetchall()
    print(result)

    sql = """
        select * from my_titanic.full_passengers LIMIT 10;
    """
    Cursor.execute(sql)

    # 取得所有資料
    result = Cursor.fetchall()
    pprint(result)
```

指定資料庫

```python
import pymysql
connection = pymysql.connect(
    host='localhost',
    user='root', 
    password='P@ssw0rd',
    port=3306,
    # 連線時指定資料庫名稱
    database="my_titanic"
)

# 建立游標
with connection.cursor() as cursor:
    # 使用游標執行 SQL 語句
    sql = "SHOW DATABASES;"
    cursor.execute(sql)
    # 取得所有資料
    databases = cursor.fetchall()

print(databases)
```

使用 cursorclass 修改回傳資料格式

```python
# 資料庫連線
connection = pymysql.connect(
    host='localhost',
    user='root', 
    password='P@ssw0rd',
    port=3306,
    database="my_titanic", # 指定database
    # 使用字典型游標
    cursorclass = pymysql.cursors.DictCursor
)

# 建立游標
with connection.cursor() as cursor:
    # 使用游標執行 SQL 語句
    sql = "SHOW DATABASES;"
    cursor.execute(sql)
    # 取得所有資料
    databases = cursor.fetchall()

print(databases)
```

查詢資料庫

```python
with connection.cursor() as cursor:
    # 使用游標執行 SQL 語句
    sql = "select * from my_titanic.full_passengers LIMIT 10;"
    cursor.execute(sql)
    # 取得所有資料
    result = cursor.fetchall()

pprint(result, sort_dicts=False)
```
