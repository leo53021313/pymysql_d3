```python
import pymysql
from configparser import ConfigParser

config = ConfigParser()
config.read('../Chapter1/config.ini')

connection = pymysql.connect(
    host=config.get('DB', 'host'),
    user=config.get('DB', 'user'),
    password=config.get('DB', 'password'),
    port=config.getint('DB', 'port'),
    cursorclass=pymysql.cursors.DictCursor,
)

# 建立資料庫
with connection.cursor() as cursor:
    sql = """
    CREATE DATABASE IF NOT EXISTS chapter3;
    """
    cursor.execute(sql)

    # 建立有時間紀錄的表格
    # 包含 last_login 欄位 預設在寫入時使用現在的時間
    sql = """
        CREATE TABLE IF NOT EXISTS chapter3.user (
            id INT AUTO_INCREMENT PRIMARY KEY,
            name VARCHAR(255) NOT NULL,
            age INT,
            username VARCHAR(255) NOT NULL UNIQUE,
            password VARCHAR(255) NOT NULL,
            last_login TIMESTAMP DEFAULT CURRENT_TIMESTAMP
        );
    """

    cursor.execute(sql)
    cursor.execute("SHOW TABLES FROM chapter3;")
    result = cursor.fetchall()

print(result)
```

```python
from pprint import pprint

connection = pymysql.connect(
    host=config.get('DB', 'host'),
    user=config.get('DB', 'user'),
    password=config.get('DB', 'password'),
    port=config.getint('DB', 'port'),
    cursorclass=pymysql.cursors.DictCursor,
    database='chapter3'
)

with connection.cursor() as cursor:
    sql = """
        INSERT INTO user(name, age, username, password)
        VALUES('Happy2', 18, 'happy02', '123456')
    """
    cursor.execute(sql)
    connection.commit()
    cursor.execute("SELECT * FROM user;")
    result = cursor.fetchall()
    
pprint(result)
```

```python
from datetime import datetime

# 取得當前時間
print(datetime.now())

# 取得指定時間的 datetime 物件
print(datetime(2025, 1, 1, 12, 00, 00))


# 輸出成指定格式
now = datetime.now()
print(now)
print(now.strftime("%Y-%m-%d %H:%M:%S"))
print(now.strftime("%Y-%m-%d"))
print(now.strftime("%Y"))

# 取得民國年
print("="*50)
now = datetime.now()
print(now.year)
print(now.year-1911)
print(now.month)
print(now.day)
print(now.date())

```

```python
# 將指定格式的字串轉換成 datetime 物件
date_time = "2026/11/12 23:11:11"

# 轉換成 datetime 物件
data = datetime.strptime(date_time, "%Y/%m/%d %H:%M:%S")
print(data)
print(data.year)
```

```python
from datetime import timedelta
# 取得前一天日期
yesterday = datetime.now() - timedelta(days=1)
print(yesterday)
print(yesterday.date())

#去年
last_year = datetime.now() - timedelta(days=365)
print(last_year.year)

```

```python
# ISO format
iso_string = '2025-01-01T12:00:00'

# 轉換成 datetime 物件
obj = datetime.fromisoformat(iso_string)
print(obj)
```
