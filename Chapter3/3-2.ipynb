```python
connection = pymysql.connect(
    host=config.get('DB', 'host'),
    user=config.get('DB', 'user'),
    password=config.get('DB', 'password'),
    port=config.getint('DB', 'port'),
    cursorclass=pymysql.cursors.DictCursor,
    db='chapter3',
)

# 建立有時間紀錄的表格
with connection.cursor() as cursor:
    sql = """
        CREATE TABLE IF NOT EXISTS users2 (
            id INT AUTO_INCREMENT PRIMARY KEY,
            name VARCHAR(255) NOT NULL,
            age INT NOT NULL,
            username VARCHAR(255) NOT NULL,
            password VARCHAR(255) NOT NULL,
            start_date DATE DEFAULT NULL,
            last_login TIMESTAMP DEFAULT CURRENT_TIMESTAMP
        );
    """
    cursor.execute(sql)
```

```python
# 取得當前時間並插入資料
from datetime import datetime

with connection.cursor() as cursor:
    sql = """
        INSERT INTO users2 (name, age, username, password, start_date)
        VALUES ("Jarvis", 25, "jarvis", "123456", %s);
    """
    # 這裡使用標準 SQL 日期格式 'YYYY-MM-DD HH:MM:SS' 或是 datetime 物件都可以
    cursor.execute(sql, (datetime.now().date()))

    connection.commit()
```

```python
with connection.cursor() as cursor:
    sql = """
        INSERT INTO users2 (name, age, username, password, start_date)
        VALUES ("Jarvis2", 25, "jarvis2", "123456", %s);
    """
    # 這裡使用標準 SQL 日期格式 'YYYY-MM-DD HH:MM:SS' 或是 datetime 物件都可以
    cursor.execute(sql, ("2026-03-11"))

    connection.commit()
```

```python
from pprint import pprint
with connection.cursor() as cursor:
    sql = """
        SELECT * FROM users2;
    """
    cursor.execute(sql)
    r = cursor.fetchall()
    pprint(r)
```
