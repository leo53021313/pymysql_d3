```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://store.line.me/stickershop/product/23398216/zh-Hant'

r = requests.get(url)

print(r)

soup = bs(r.text, 'lxml')
```

```python
import json
import os

folder = 'line_stickers'
if not os.path.exists(folder):
    os.mkdir(folder)

elements = soup.select("ul.mdCMN09Ul.FnStickerList > li")
print(len(elements))

n = 0
for element in elements:
    print(element)
    img_element = element['data-preview']
    print(img_element)
    img_url = json.loads(img_element)['staticUrl']
    print(img_url)
    #print(img_url['staticUrl']) # 會報錯因為回傳的是string 而不是 dict
    
    n += 1
    with open(f'{folder}/sticker_{n}.png', mode = 'wb') as img:
        img.write(requests.get(img_url).content)
```
