```python
from sql_utils import WebScrapingDB
db = WebScrapingDB(database=None)
db.create_database()
db = WebScrapingDB()
db.create_table()
```

```python
db.insert_store("PChome", "https://www.pchome.com.tw/")
db.insert_store("Momo", "https://www.momoshop.com.tw/")
```

```python
from commodity import ProductScrap
from concurrent.futures import ThreadPoolExecutor

product_name = "ROG滑鼠"
scrap = ProductScrap()
with ThreadPoolExecutor(max_workers=2) as executor:
    future_pchome = executor.submit(scrap.get_pchome_products, product_name)
    future_momo = executor.submit(scrap.get_momo_products, product_name)

    pc_home_products = future_pchome.result()
    momo_products = future_momo.result()

print(f"PChome 共找到 {len(pc_home_products)} 筆商品資料")
print(f"Momo 共找到 {len(momo_products)} 筆商品資料")
```
